> - Report these columns and the number of outliers detected.

### Jawaban No 4

```python
# split data
df_num = df.select_dtypes(exclude = ['object'])
```

```python
# detecting outliers using boxplot
for i in df_num:
    sns.boxplot(data = df_num, x=i)
    plt.show()
```

```python
# mengidentiifkasi outliers menggunakan IQR
def find_outliers_iqr(column):
    if df[column].dtype in ['int64', 'float64']:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        return outliers
    else:
        return pd.DataFrame()
```

```python
# mengindentifikasi kolom dengan outliers
columns_with_outliers = []
outliers_counts = {}

for column in df.columns:
    outliers = find_outliers_iqr(column)
    if not outliers.empty:
        columns_with_outliers.append(column)
        outliers_counts[column] = len(outliers)
        
# menghapus baris yg memiliki outliers
for column in columns_with_outliers:
    outliers = find_outliers_iqr(column)
    df = df.drop(outliers.index)
```

```python
# menampilkan kolom dengan outliers dan jumlah outliers
print("kolom dengan outliers dan jumlah outliers yang terdeteksi: ")
for column in columns_with_outliers:
    print(f"{column}: { outliers_counts[column]} outliers")
```

```python
# menampilkan after handling outliers 
print("Number of rows after handling outliers:", len(df))
```

### Jawaban No 5

> **Descriptive Statistics**

```python
# Central Tendencies for Numerical Columns
numerical_columns = df.select_dtypes(include=['int64', 'float64'])
central_tendencies = numerical_columns.agg(['mean', 'median', lambda x: x.mode().iloc[0]])

# Dispersion for Numerical Columns
dispersion = numerical_columns.agg(['var', 'std', lambda x: x.max() - x.min()])

# Distribution of Categorical Variables
categorical_columns = df.select_dtypes(include=['object'])
categorical_distribution = categorical_columns.apply(pd.Series.value_counts)

# Print the results
print("Central Tendencies for Numerical Columns:")
print(central_tendencies)

print("\nDispersion for Numerical Columns:")
print(dispersion)

print("\nDistribution of Categorical Variables:")
print(categorical_distribution)
```

> **Visual Analysis**

```python
# Visual Analysis - Histograms for numerical columns
plt.figure(figsize=(12, 8))
for column in numerical_columns.columns:
    sns.histplot(df[column], kde=True, label=column, bins=20)
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histograms of Numerical Columns")
plt.show()

# Visual Analysis - Box Plot for numerical columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=numerical_columns, orient="v", palette="Set2")
plt.title("Box Plot of Numerical Columns")
plt.show()

# Visual Analysis - Bar Chart for a categorical variable (replace 'categorical_column' with your actual column)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Ticket', palette="Set2")
plt.title("Distribution of Categorical Variable")
plt.xticks(rotation=45)
plt.show()
```
